```python
import pandas as pd 
from pathlib import Path 
from statsmodels.stats import multitest
import numpy as np
```

```python
wkdir="/lustre/scratch126/humgen/projects/interval_rna/interval_rna_seq/thomasVDS/misexpression_v3"
wkdir_path = Path(wkdir)

outdir_path = wkdir_path.joinpath("4_vrnt_enrich/enrich_results_af50")
outdir_path.mkdir(parents=True, exist_ok=True)
```

```python
all_enrich_results_path = wkdir_path.joinpath("4_vrnt_enrich/enrich_results_af50/snp_indel_sv_all_enrich_results_z_cutoff.tsv")
all_enrich_results_df = pd.read_csv(all_enrich_results_path, sep="\t")
```

```python
# SNV, indel, SV across z-score cutoffs and AF bins
z_cutoff_bins = [2, 10, 20, 30, 40]
snv_indel_sv_zscore_af_df = all_enrich_results_df[all_enrich_results_df.vrnt_type.isin(["all_sv", "snp", "indel"]) & 
                                                  (all_enrich_results_df.consequence == "all") & 
                                                  (all_enrich_results_df.z_cutoff.isin(z_cutoff_bins)) & 
                                                  (all_enrich_results_df.window_name.isin(["gene body +/-10kb", "gene body +/-200kb"]))
                                                 ]
print(f"Number of tests: {snv_indel_sv_zscore_af_df.shape[0]}")
```

```python
# Rare SNV, indel, SV across z-score cutoffs and AF bins
z_cutoff_bins = [2, 10, 20, 30, 40]
snv_indel_sv_rare_windows_zscore_df = all_enrich_results_df[all_enrich_results_df.vrnt_type.isin(["all_sv", "snp", "indel"]) & 
                                                            (all_enrich_results_df.consequence == "all") & 
                                                            (all_enrich_results_df.maf_range == "0-1") &
                                                            (all_enrich_results_df.z_cutoff.isin(z_cutoff_bins)) & 
                                                            ~(all_enrich_results_df.window_name.isin(["gene body +/-10kb", "gene body +/-200kb"]))
                                                 ]
print(f"Number of tests: {snv_indel_sv_rare_windows_zscore_df.shape[0]}")
```

```python
# rare DEL, DUP, INV, MEI SV z-score cutoffs 
sv_types_rare_zscore_df = all_enrich_results_df[all_enrich_results_df.vrnt_type.isin(["DEL", "DUP", "INV", "MEI"]) &
                                                (all_enrich_results_df.consequence == "all") & 
                                                (all_enrich_results_df.z_cutoff.isin(z_cutoff_bins)) & 
                                                (all_enrich_results_df.maf_range == "0-1") &
                                                (all_enrich_results_df.window_name.isin(["gene body +/-10kb", "gene body +/-200kb"]))]
print(f"Number of tests: {sv_types_rare_zscore_df.shape[0]}")
```

```python
# rare DEL, DUP, INV, MEI SV z-score cutoffs, windows  
sv_types_rare_windows_zscore_df = all_enrich_results_df[all_enrich_results_df.vrnt_type.isin(["DEL", "DUP", "INV", "MEI"]) &
                                                (all_enrich_results_df.consequence == "all") & 
                                                (all_enrich_results_df.z_cutoff.isin(z_cutoff_bins)) & 
                                                (all_enrich_results_df.maf_range == "0-1") &
                                                ~(all_enrich_results_df.window_name.isin(["gene body +/-10kb", "gene body +/-200kb"]))]
print(f"Number of tests: {sv_types_rare_windows_zscore_df.shape[0]}")
```

```python
#sv_msc_enrich_results_path = wkdir_path.joinpath("4_vrnt_enrich/enrich_results_sv_gene_msc/sv_msc_by_gene_all_enrich_results.tsv")
#sv_msc_enrich_results_df = pd.read_csv(sv_msc_enrich_results_path, sep="\t")
```

```python
# remove consequences never seen across the dataset 
sv_msc_enrich_zscore_2_df = all_enrich_results_df[all_enrich_results_df.vrnt_type.isin(["DEL", "DUP", "INV", "MEI"]) &
                                                     (all_enrich_results_df.consequence != "all") & 
                                                     (all_enrich_results_df.z_cutoff == 2) & 
                                                     (all_enrich_results_df.maf_range == "0-1")].copy()
sv_msc_enrich_zscore_2_df["total_carrier"] = sv_msc_enrich_zscore_2_df.misexp_carrier + sv_msc_enrich_zscore_2_df.control_carrier
sv_msc_zscore_2_total_carriers_df = sv_msc_enrich_zscore_2_df[["vrnt_type", "consequence", "total_carrier"]]
sv_msc_nonzero_carrier_df = sv_msc_zscore_2_total_carriers_df[sv_msc_zscore_2_total_carriers_df.total_carrier != 0]
sv_msc_nonzero_carrier_df = sv_msc_nonzero_carrier_df.drop(columns=["total_carrier"])
print(f"SV consequences with carriers: {sv_msc_nonzero_carrier_df.shape[0]}")
```

```python
sv_msc_enrich_zscore_df = all_enrich_results_df[all_enrich_results_df.vrnt_type.isin(["DEL", "DUP", "INV", "MEI"]) &
                                            (all_enrich_results_df.consequence != "all") & 
                                            (all_enrich_results_df.z_cutoff.isin(z_cutoff_bins)) & 
                                            (all_enrich_results_df.maf_range == "0-1")]


sv_msc_enrich_zscore_2_nonzero_carrier_df = pd.merge(sv_msc_enrich_zscore_df, 
                                                    sv_msc_nonzero_carrier_df, 
                                                    on=["vrnt_type", "consequence"], 
                                                    how="inner")
print(f"Number of tests: {sv_msc_enrich_zscore_2_nonzero_carrier_df.shape[0]}")
```

```python
all_enrich_tests_df_list = [snv_indel_sv_zscore_af_df, 
                            snv_indel_sv_rare_windows_zscore_df,
                            sv_types_rare_zscore_df,
                            sv_types_rare_windows_zscore_df, 
                            sv_msc_enrich_zscore_2_nonzero_carrier_df
                           ]
all_enrich_tests_df = pd.concat(all_enrich_tests_df_list).drop(columns=["consequence_name"])
test_num = all_enrich_tests_df.shape[0]
print(f"Total number of tests: {all_enrich_tests_df.shape[0]}")
```

```python
### multiple testing correction 
# multiple testing correction (BH FDR method)
pval_as_array = all_enrich_tests_df.pval.to_numpy()
for method in ["fdr_bh", "bonferroni"]: 
    pass_test, pval_adj, _, _ = multitest.multipletests(pval_as_array, alpha=0.05, method=method)
    all_enrich_tests_df[f"{method}_pass"] = pass_test
    all_enrich_tests_df[f"{method}_pval_adj"] = pval_adj
```

```python
# write to file
all_enrich_tests_bonf_path = outdir_path.joinpath(f"snp_indel_sv_all_enrich_results_af50_{test_num}_bonf_adj.tsv")
all_enrich_tests_df.to_csv(all_enrich_tests_bonf_path, sep="\t", index=False)
```
